We need to format RAM and Storage and extract to store it as GB

```python
df['GPU'].unique()
```

```python
class DataCleaner:
    """
    Data Cleaning helper class
    """
    @staticmethod
    def clean_ram(x):
        ram = re.findall(r'\d+', x)
        return int(ram[0])

    def transform_brand(x):
        # function to select popular brand and mark others as 'other'
        if x.lower() in ['asus', 'lenovo', 'hp', 'msi', 'dell', 'acer', 'apple', 'microsoft', 'toshiba', 'samsung']:
            return x.lower()
        return 'other'

    def transform_storage(x):
        storage = x.split(' ')[0]
        if 'GB' in storage:
            s = re.findall(r'\d+', storage)
            return int(s[0])
        elif 'TB' in storage:
            s = re.findall(r'\d+', storage)
            return int(s[0]) * 1000
        else:
            return int(storage)

    def transform_gpu(x):
        if type(x) == str:
            x = x.lower()
            if x == 'intel':
                return 'intel'
            elif x == 'nvdia':
                return 'nvdia'
            elif x == 'amd' or 'radeon' in x:
                return 'amd'
            elif x == 'other':
                return 'other'
            return 'nvdia'
        else:
            return 'intel'
        
        
```

```python
df['RAM'] = df['RAM'].apply(DataCleaner.clean_ram)
df['Brand'] = df['Brand'].apply(DataCleaner.transform_brand)
df['Storage'] = df['Storage'].apply(DataCleaner.transform_storage)
df['GPU'] = df['GPU'].apply(DataCleaner.transform_gpu)
```
