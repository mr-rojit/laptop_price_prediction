```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
```

```python
product_list = []
product_price_list = []
```

```python
try:
    for i in range(1, 401):
        full_url = f"https://www.amazon.com/s?rh=n%3A565108&fs=true&page={i}&qid=1712594591&ref=sr_pg_{i}"
        response=requests.get(full_url, headers={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'})

        # Check if the request was successful
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')

            # Use a dictionary of attributes in find_all()
            product_info = soup.find_all('span', {'class': 'a-size-base-plus a-color-base a-text-normal'}, limit=33)
            product_price = soup.find_all('span', {'class': 'a-offscreen'}, limit=33)

            # Use list comprehension to simplify appending to lists
            product_list.extend([product for product in product_info])
            product_price_list.extend([price.text for price in product_price])
        else:
            print(f'Request failed with status code {response.status_code}')
except requests.exceptions.RequestException as e:
    print(f'Error connecting to Amazon: {e}')
```

```python
df = pd.read_csv('product_list_amazon_final.csv', index_col=0)
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df['CPU'] = df['Product_info'].str.extract('(Intel|AMD) [^\,]*')
df['RAM'] = df['Product_info'].str.extract('(\d+GB RAM)')
df['Storage'] = df['Product_info'].str.extract('(\d+GB SSD|\d+TB HDD)')
df['Screen_size'] = df['Product_info'].str.extract('(\d+\.\d+")') #[^\,]*
df['Brand'] = df['Product_info'].apply(lambda x: x.split()[0])
df['OS'] = df['Product_info'].str.extract('(Windows \d+|Mac OS|Linux)')
df['Gaming'] = df.apply(lambda row: 'Yes' if 'Gaming' in row.Product_info else 'No', axis=1)
df["GPU"] = df["Product_info"].apply(lambda s: "Yes" if "NVIDIA" in s or "AMD Radeon" in s else "No")

print(df[['CPU', 'RAM', 'Storage', 'Screen', 'Brand', 'OS', 'Gaming']])
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.amazon.com/A315-24P-R7VH-Display-Quad-Core-Processor-Graphics/dp/B0BS4BP8FB/ref=sr_1_4?s=pc&sr=1-4"

response=requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'})
print(response)
soup = BeautifulSoup(response.content, "html.parser")

#table = soup.find("table", {"class": "a-keyvalue prodDetTable"})
table = soup.find("table", {"id": "productDetails_techSpec_section_1"})

laptops_data = []

for row in table.find_all("tr"):
    columns = row.find_all("td")
    if len(columns) >= 2:  
        laptop_name = columns[0].text.strip()
        laptop_description = columns[1].text.strip()
        laptops_data.append({"Name": laptop_name, "Description": laptop_description})

df_info = pd.DataFrame(laptops_data)
df_info
```
