```python
class DataCleaner:
    """
    Data Cleaning helper class
    """
    @staticmethod
    def clean_ram(x):
        ram = re.findall(r'\d+', x)
        return int(ram[0])

    def transform_brand(x):
        # function to select popular brand and mark others as 'other'
        if x.lower() in ['asus', 'lenovo', 'hp', 'msi', 'dell', 'acer', 'apple', 'microsoft', 'toshiba', 'samsung']:
            return x.lower()
        return 'other'

    def transform_storage(x):
        storage = x.split(' ')[0]
        if 'GB' in storage:
            s = re.findall(r'\d+', storage)
            return int(s[0])
        elif 'TB' in storage:
            s = re.findall(r'\d+', storage)
            return int(s[0]) * 1000
        else:
            return int(storage)

    def transform_gpu(x):
        if type(x) == str:
            x = x.lower()
            if x == 'intel':
                return 'intel'
            elif x == 'nvdia':
                return 'nvdia'
            elif x == 'amd' or 'radeon' in x:
                return 'amd'
            elif x == 'other':
                return 'other'
            return 'nvdia'
        else:
            return 'intel'
        
        
```

```python
df['RAM'] = df['RAM'].apply(DataCleaner.clean_ram)
df['Brand'] = df['Brand'].apply(DataCleaner.transform_brand)
df['Storage'] = df['Storage'].apply(DataCleaner.transform_storage)
df['GPU'] = df['GPU'].apply(DataCleaner.transform_gpu)
```

```python
df['GPU'].value_counts()
```

```python
## transform processor

def transform_cpu(x):
    if type(x) == str:
        x = x.lower()
        if 'intel' in x and 'i7' in x:
            return 'i7'
        elif 'intel' in x and 'i5' in x:
            return 'i5'
        elif 'intel' in x and 'i3' in x:
            return 'i3'
        elif 'intel' in x and 'i5' in x:
            return 'i5'
        elif '9' in x:
            return 'i9'
        elif 'ryzen' in x and '7' in x:
            return 'ryzen 7'
        elif 'ryzen' in x and '5' in x:
            return 'ryzen 5'
        elif 'ryzen' in x and '3' in x:
            return 'ryzen 3'
        elif 'apple' in x and '2' in x:
            return 'M2'
        elif 'apple' in x and '1' in x:
            return 'M1'
        return 'other'
    else:
        return 'ryzen 5'
```

```python
df['CPU'] = df['CPU'].apply(transform_cpu)
```

```python
def storage_quantile_capping(data):
    lq = data.quantile(0.02)
    uq = data.quantile(0.99)
    capped_data = data.clip(lq, uq)
    return capped_data

def ram_quantile_capping(data):
    lq = data.quantile(0.01)
    uq = data.quantile(0.99)
    capped_data = data.clip(lq, uq)
    return capped_data
    
```
